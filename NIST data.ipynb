```python
# filename = 'NIST/74-82-8-Mass.jdx'
filename = 'NIST/7732-18-5-Mass.jdx'

A = np.array([])
with open(filename) as f:
    counter=0
    for x in f:
        counter=counter+1
        if (counter>9) & (x[0] !=("#" or "C"or "o")):
            xp = x.replace(b' ',b',') 
#             print xp
            A=np.append(A,np.loadtxt( StringIO.StringIO(xp), delimiter=','))
m= A[0::2]  
r= A[1::2]
```
